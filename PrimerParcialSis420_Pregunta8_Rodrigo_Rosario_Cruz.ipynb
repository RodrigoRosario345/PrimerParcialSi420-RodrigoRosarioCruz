```python
'''
correcion reemplazando cerillos de numeros.

la solucion se basara en usar tipo de variables o almacenar enteros o en caso de las los signos +  y - se lo almacenara como medidas de cambios
"int" "+ or -" "int" "=" "int"
donde tendremos valores de rango entre 0-9
ejemplo "4-1=6"
'''

#import pdb
#pdb.set_trace()

def Solucion(num,mapa_fosforos,todos, mas_uno, cero, menos_uno):
  print("El estado inicio es: ",num)
  primero = int(num[0])
  signo = num[1]
  segundo = int(num[2])
  tercero = int(num[4])
  
  cerillos_totales = mapa_fosforos[primero]+mapa_fosforos[signo]+mapa_fosforos[segundo]+mapa_fosforos[tercero]

  elecciones_primero = todos[primero]
  elecciones_segundo = todos[segundo]
  elecciones_tercero = todos[tercero]
  elecciones_signo = ["+","-"]
  
  posible  = []
  agregar = 0
  mover = 0
  eliminar = 0

#recorrer y agregar d
  for i in elecciones_primero:
    for j in elecciones_segundo:
      for k in elecciones_tercero:
        for l in elecciones_signo:
          cerillos = mapa_fosforos[i]+mapa_fosforos[j]+mapa_fosforos[k]+mapa_fosforos[l]
          if cerillos == cerillos_totales:

            try:
              if i in mas_uno[primero]: agregar+=1
            except: 
              pass
            try:
              if i in cero[primero]: mover+=1
            except: 
              pass
            try:
              if i in menos_uno[primero]: eliminar+=1
            except: 
              pass
            try:
              if j in mas_uno[segundo]: agregar+=1
            except: 
              pass
            try:
              if j in cero[segundo]: mover+=1
            except: 
              pass
            try:
              if j in menos_uno[segundo]: eliminar+=1
            except: 
              pass
            try:
              if k in mas_uno[tercero]: agregar+=1
            except: 
              pass
            try:
              if k in cero[tercero]: mover+=1
            except: 
              pass
            try:
              if k in menos_uno[tercero]: eliminar+=1
            except: 
              pass
            try:
              if l in mas_uno[signo]: agregar+=1
            except: 
              pass
            try:
              if l in menos_uno[signo]: eliminar+=1
            except: 
              pass
            #print(agregar, eliminar, [i,l,j,k])
            if agregar==1 and eliminar==1 and mover==0:              
              posible.append([i,l,j,k]) 
            elif mover==1 and eliminar==0 and agregar==0:
              posible.append([i,l,j,k]) 
            agregar=0
            eliminar=0
            mover=0
  #print("Soluciones posibles : ",posible)     
  correcto=0      
  for i in posible:
    #print(i)
    resultado = eval(str(i[0])+i[1]+str(i[2]))
    if resultado==i[3]:
      correcto+=1
      print("La Solucion Correcta es : "+str(i[0])+i[1]+str(i[2])+"="+str(i[3]))
  if correcto==0:
    print("No existe solucion para esos numeros : ",num)
  return None
      

mapa_fosforos = {0:6,1:2,2:5,3:5,4:4,5:5,6:6,7:3,8:7,9:6,"+":2,"-":1}

mas_uno = {0:[8],1:[7],2:[],3:[],4:[],5:[6,9],6:[8],7:[],8:[],9:[8],"-":["+"]}
cero = {0:[6,9],1:[],2:[3],3:[5,2],4:[],5:[3],6:[0,9],7:[],8:[],9:[0,6]}
#no se puede = {4:4}
menos_uno = {0:[],1:[],2:[],3:[],4:[],5:[],6:[5],7:[1],8:[0,9,6],9:[5,3],"+":["-"]}

todos = {0:[0,6,9,8],1:[1,7],2:[2,3],3:[3,5,2],4:[4],5:[5,3,6,9],6:[6,0,5,8,9],7:[7,1],8:[8,0,6,9],9:[9,0,3,5,6,8]}

estado_inicio = []
estado_inicio.append("5+7=2") # 9 - 7 = 2

estado_inicio.append("4-1=6") # 4 + 1 = 5
estado_inicio.append("6+4=4") #  0 + 4 = 4 y 8 - 4 = 4
estado_inicio.append("6-2=1")  # no existe solucion
estado_inicio.append("6+8=8") # 8 + 0 = 8 o 0 + 8 = 8
estado_inicio.append("3+3=5") # 3+2=5 o 5+2=3
estado_inicio.append("5+9=9") 

for i in estado_inicio:
  Solucion(i, mapa_fosforos, todos, mas_uno, cero, menos_uno)
  
  
```
